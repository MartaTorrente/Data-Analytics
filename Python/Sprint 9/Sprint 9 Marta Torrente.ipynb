### 1. Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en: 🔗 Índice de masa corporal IMC que es y como se calcula.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def calculo_IMC():

    """ Función para calcular el Índide de Masa Corporal (IMC)
         a partir del peso y la altura introducidos por el usuario
    """

    while True:  # repite hasta que los datos introducidos sean válidos
        try: 
            peso = float(input('Introduce tu peso en kg').replace(',','.'))
            altura = float(input('Introduce tu altura en metros').replace(',','.'))

            if peso < 0 or altura <=0:      # validamos que peso y altura sean valores positivos
                print('Por favor, introduce un valor positivo')
                continue
            if peso >250 or altura > 3:     # validamos rango de parámtros dentro de normalidad
                print('Por favor, introduce valores dentro de la normalidad')
                continue

            imc = round((peso/altura**2), 2)          # cálculo del imc

            if imc < 18.5:                  # clasificamos según el valor de IMC
                print('Tu IMC es de:', imc, 'Estás en Bajo peso')
            elif imc  < 25:
                print('Tu IMC es de:', imc,'Estás en peso normal')
            elif imc < 30:
                print('Tu IMC es de:', imc,'Estás en sobrepeso')
            else:
                print('Tu IMC es de:', imc,'Estás en obesidad')
            break
        except ValueError:                  # se ejecuta si el usuario introduce valores no numéricos
            print('Por favor, Introduce números válidos')


calculo_IMC() 
```

### 2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
def convertidor_temperatura():

    """ Convierte la tempreatura introducida por el usuario desde una medida origen 
        al resto de medidas disponibles (Celsius, Farenheit y Kelvin)
    """

    # Diccionario para todas las conversiones posibles
    conversiones = {
        'C': {
            'Farenheit': lambda t:(t*9/5)+32,
            'Kelvin': lambda t: t +273.15
        },
        'F':{
            'Celsius': lambda t : (t - 32) * 5/9,
            'Kelvin': lambda t: ((t - 32) * 5/9) + 273.15
        },
        'K': {
            'Celsius': lambda t: t - 273.15,
            'Farenheit': lambda t: ((t - 273.15) * 9/5) + 32
        }

    }

    while True:     # repite hasta que los datos introducidos sean válidos
        try:
            unidad = input('Indica la unidad de medida que quieres usar: C/F/K').upper()
            temperatura = float(input('Introduce la temperatura'))

            if unidad not in conversiones:      # Valida que la unidad exista en el diccionario
                print('La unidad introducida no es válida')
                continue

            resultados = {}

            for destino, formula in conversiones[unidad].items():   # Recorre las conversiones disponibles para esa unidad donde la clave es el destino y el valor la fórmula
                resultados[destino] = formula(temperatura)           # Aplica la fórmula que corresponda

            return {clave: round(valor, 2) for clave, valor in resultados.items()}  #extrae cada clave:valor del diccionario, redondea el valor y devuelve           


        except:       # se ejecuta si el usuario introduce valores no válidos
            print('Debes introducir números válidos')


convertidor_temperatura()

```

### 3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
# se importa el módulo re (regular expresion)
import re 
```

```python
def contador_palabras_texto():
    """
    Función para contar cuántas veces aparece
    cada palabra en un texto introducido por el usuario.
    """

    texto = input('Introduce el texto: ').lower()

    
    palabras = re.findall(r'\b\w+\b', texto)        # Extrae palabras ignorando signos de puntuación

    contador = {}

    for palabra in palabras:
        if palabra not in contador:
            contador[palabra] = 1
        else:
            contador[palabra] += 1

    
    if len(palabras) > 0:       # Evita división por cero 
        media_longitud_palabras = (
            sum(len(palabra) for palabra in palabras)
            / len(palabras)
        )
    else:
        media_longitud_palabras = 0

    return contador, round(media_longitud_palabras, 2)


contador, media = contador_palabras_texto()

print("Contador de palabras:", contador, "\nMedia de longitud de palabras:", media)
```

### 4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

```python
def invertir_diccionario(diccionario):

    """"
    Función que crea un nuevo diccionario 
    invirtiendo el orden del diccionario original
    
    """
    
    invertido = {}

    for clave, valor in diccionario.items(): # recorre cada elemento del diccionario y extrae clave y valor

        if valor not in invertido:

            invertido[valor] = [clave]

        else:

            invertido[valor].append(clave)

   
    for valor, claves in invertido.items():

        if len(claves) > 1: # comprueba si hay más de una clave para el valor
            
            print('Advertencia: valor duplicado', [valor]) 
            print('Asociado a:',claves)

    return invertido

invertir_diccionario({ "Ana": 45,
    "Luis": 30,
    "Marta": 45})
```

### 1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. 

```python
import re
from pprint import pprint

def contador_organizador_palabras(archivo):

    """
    Función que abre archivo y cuenta palabras y las ordena alfabéticamente
    según su letra inicial.
    """

   
    texto = archivo.read().lower()   

    palabras = re.findall(r'\b\w+\b', texto)     # Extrae palabras del texto 

    contador = {}

    for palabra in palabras:    

        inicial = palabra[0] # Obtener letra posición inicial

        if inicial not in contador: # Si la letra no existe, crear subdiccionario
            contador[inicial] = {}

        if palabra not in contador[inicial]:     # Cuenta las palabras
            contador[inicial][palabra] = 1
        else:
            contador[inicial][palabra] += 1

  
    contador_ordenado = {}

    for letra in sorted(contador):    # Ordena el diccionario

        contador_ordenado[letra] = dict(sorted(contador[letra].items())) # ordena el diccionario de cada letra, por su clave

    return contador_ordenado

with open('tu me quieres blanca.txt', 'r', encoding='utf-8') as archivo: 


 resultado = contador_organizador_palabras(archivo)


pprint(resultado)

```

### 2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
def separador_datos(datos):
    """
        Función que permite separa elementos de una lista
        en función de si se pueden convertir en flotantes o no
    """
    normalizada = []  # lista vacía para guardar los datos normalizados sin listas o tuplas

    for elemento in datos:
        if isinstance(elemento,(list, tuple)):  # se comprueba si el elemento es una lista o tupla
            for subelemento in elemento:          # si es así, recorre cada subelemento y se añade a la lista
                normalizada.append(subelemento)
        else:                                      # si no es lista o tupla añade el elemento
            normalizada.append(elemento)

            
    flotantes = []
    no_flotantes = []

    for item in normalizada:       
        try:
            flotantes.append(float(item))   #se intenta la conversión a float y se añade a la lista
        except:                             # si no se puede convertir se añada a la otra lista
            no_flotantes.append(item)
    return flotantes, no_flotantes

datos = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

flotantes, no_flotantes = separador_datos(datos)

print('Flotantes:',flotantes, '\nNo flotantes:', no_flotantes )
```

### 1. Generador de Contrasenyes

Explora el funcionament del mòdul random de la llibreria numpy. 🔗 Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

- longitud (int): Longitud de la contrasenya
- majuscules (bool = True): Si hi ha d’aparèixer majúscules
- minuscules (bool = True): Si hi ha d’aparèixer minúscules
- numeros (bool = True): Si hi ha d’aparèixer números
- signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python

import numpy as np
import string  # módulo con colecciones de caracteres ya preparadas
import pyperclip # módulo que permite copiar al portapapeles y pegar desde el portapapeles


def crear_contrasenya(
    longitud,
    majuscules=True,
    minuscules=True,
    numeros=True,
    signes=False
):

    """
        Función que permite crear contraseñas 
        de forma aleatoria según paramentros especificados
    """

    caracteres = ""     # se irán guardando todos los caractares
    contrasenya = []    # lista donde se guardaran los caracteres generados

    if majuscules:
        caracteres += string.ascii_uppercase  # colección de abecedario en mayúsculas que añadimos a  caracteres
        contrasenya.append(np.random.choice(list(string.ascii_uppercase))) # añadimos a la lista de forma aleatoria una mayúscula

    if minuscules:
        caracteres += string.ascii_lowercase    # colección de abecedario en minúsculas
        contrasenya.append(np.random.choice(list(string.ascii_lowercase)))

    if numeros:
        caracteres += string.digits         # colección de numeros del 0-9
        contrasenya.append(np.random.choice(list(string.digits)))

    if signes:
        caracteres += string.punctuation    # colección de caracteres especiales
        contrasenya.append(np.random.choice(list(string.punctuation)))

    caracteres = list(caracteres)    # se convierte a lista

    while len(contrasenya) < longitud:   # mientras la contraseña sea más corta que la longitud indicada se añaden aleatoriamente caracteres
        contrasenya.append(np.random.choice(caracteres))

    np.random.shuffle(contrasenya)   # se mezcla el orden de los caracteres aleatoriamente

    
    contrasenya_final = "".join(contrasenya)  # # Convertimos lista -> string, join permite unir elementos de una lista


    pyperclip.copy(contrasenya_final)      # Copiamos al portapapeles

    return contrasenya_final


print(crear_contrasenya(10, True, True, True, True), '--> copiada en portapapeles')
```

### 2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l’equip més golejador.
El nom de l’equip més golejat
La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)

Pots utilitzar el format que et sigui més convenient.

```python
goles_a_favor = {}
goles_en_contra = {}
clasificacion = {}

with open('historic partits.txt', 'r', encoding='utf-8') as archivo:

    for linea in archivo:

        linea = linea.strip()  # elimina espacios en blanco, tabulaciones del inicio y final de linea
        equipo1, marcador, equipo2 = linea.split('\t')  # separa el texto según tabulación
        goles1, goles2 = marcador.split('-')  # Separa los goles por el guión
       
        goles1 = int(goles1)     # Convierte a enteros
        goles2 = int(goles2)

        for equipo in [equipo1, equipo2]:     

            if equipo not in goles_a_favor:
                goles_a_favor[equipo] = 0

            if equipo not in goles_en_contra:
                goles_en_contra[equipo] = 0

            if equipo not in clasificacion:
                clasificacion[equipo] = 0

        
        goles_a_favor[equipo1] += goles1  # cuenta goles a favor
        goles_a_favor[equipo2] += goles2

        
        goles_en_contra[equipo1] += goles2  # cuenta goles en contra
        goles_en_contra[equipo2] += goles1

        
        if goles1 > goles2:         # clasificación por puntos

            clasificacion[equipo1] += 3

        elif goles1 == goles2:

            clasificacion[equipo1] += 1
            clasificacion[equipo2] += 1

        else:

            clasificacion[equipo2] += 3

mas_goleador = max(goles_a_favor, key=goles_a_favor.get) # para indicar a la funcion max que extraiga y compare los valores de cada clave (.get)

mas_goleado = max(goles_en_contra, key=goles_en_contra.get)

clasificacion_ordenada = dict(  # convierte de nuevo en diccionario
        sorted(     # devuelve en formato lista los elementos ordenados
        clasificacion.items(), # extrae los elementos 
        key=lambda item: item[1],  # indica que se debe ordenar por valor (intem[1])
        reverse=True # ordena los elementos de formas descendiente
    )
)


print('\n' + '=' * 60)
print('CLASIFICACIÓN GLOBAL')
print('=' * 60)

# Encabezados
print(f'{"Equipo":25} {"Pts":>5} {"GF":>5} {"GC":>5}')

print('-' * 60)

# Filas
for equipo, puntos in clasificacion_ordenada.items():

    gf = goles_a_favor[equipo]
    gc = goles_en_contra[equipo]

    print(f'{equipo:25} {puntos:>5} {gf:>5} {gc:>5}')

print('=' * 60)

print(f'\nEquipo más goleador: {mas_goleador}')
print(f'Equipo más goleado: {mas_goleado}')
```
