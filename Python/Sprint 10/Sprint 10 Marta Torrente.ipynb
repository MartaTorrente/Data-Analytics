Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.

```python
df['Data Naixement'] = pd.to_datetime({
    'year' : df['Any'], 
    'month' : df['Mes'],
    'day' : df['Dia']
})


```

```python
from datetime import date

def calcular_edad(data):

    """
    Calcula la edad actual a partir de una fecha indicada
    """
    avui = date.today()   # para saber la fecha actual

    edat = avui.year - data.year  # restamos año actual - año de la fecha indicada

    if (avui.month, avui.day) < (data.month, data.day):  # verificamos si ya ha cumplido años mediante la comparación del mes y dia actual y de la fecha indicada
        edat -= 1

    return edat
```

```python
df["Data Naixement"] = pd.to_datetime(df["Data Naixement"])
df["Data Naixement"].dtype
```

Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
df['Edat actual'] = df['Data Naixement'].apply(calcular_edad)

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creamos tres variables para cada tipo de dato

columnas_numericas = df.select_dtypes(include='number').columns

columnas_fecha = df.select_dtypes(include='datetime').columns

columnas_categoricas = df.select_dtypes(
    exclude=['number', 'datetime']
).columns
```

```python
import matplotlib.pyplot as plt

def crear_graficos(df):

    """
    Crea y exporta gráficos automáticamente 
    para cada columna según su tipo del DataFrame indicado
    - Histograma si es numérica
    - Barras por año si es fecha
    - Top 10 si es categórica
    """ 

    columnas_numericas = df.select_dtypes(include='number').columns
    columnas_fecha = df.select_dtypes(include='datetime').columns

    for columna in df.columns:

        plt.figure(figsize=(6,4))  # para crear una figura nueva para cada gráfico

        if columna in columnas_numericas:

            df[columna].dropna().hist()
            plt.title('Histograma de '+columna)

        elif columna in columnas_fecha:

            df[columna].dt.year.value_counts().sort_index().plot.bar()
            plt.title('Anys de ' +columna)

        else:

            df[columna].value_counts().head(10).plot.bar()
            plt.title('Valors més freqüents de ' +columna)

        plt.tight_layout() # ajusta de forma automática el diseño

        plt.savefig(f'columna.png') # Guarda cada gráfico 

        plt.show()  # muestra el gráfico

df = sns.load_dataset("titanic")
crear_graficos(df)
```

### 2.

Carrega l'arxiu matriu_distancies.xlsx a pandas, de manera que els noms de files i els noms de columnes siguin els de les ciutats. Borra "Las Palmas de Gran Canaria" i "Palma" perquè poguem fer el trajecte en cotxe.

Font: Mejores Rutas

Ens interessa visitar totes les ciutats principals d'Espanya recorrent la mínima distància possible.

No cal que ho facis de forma òptima, ens interessa que desenvolupis una solució raonable utilitzant les eines que tens actualment.

Per exemple, una aproximació senzilla (que no òptima) seria anant sempre a la ciutat més propera que no haguem visitat encara

Fes una funció que donada la matriu de distàncies i la ciutat d'origen, faci una proposta de ruta que sigui el més curta possible que puguis, retornant una llista amb l'ordre de visita. Dóna també la distància total recorreguda.

```python
df = pd.read_excel(
    "matriu_distancies.xlsx",
    index_col=0
)

df = df.drop (
    index = ['Las Palmas de Gran Canaria', 'Palma'],
    columns = ['Las Palmas de Gran Canaria', 'Palma']
)
```

```python

def calcular_ruta(df, origen):

    """
    Calcula una ruta aproximada, devuelve la ruta y la distancia total recorrida
    dada una matriz de distáncias y una ciudad origen
    """

    ruta = [origen]  # se inicia la ruta con la ciudad origen
    actual = origen
    distancia_total = 0
    no_visitadas = set(df.columns)  # crea un conjunto con todas las ciudades
    no_visitadas.remove(origen)  # se elimina la ciudad origen

    while no_visitadas:
        siguiente = min(        # selecciona la ciudad más cercana a la actual
            no_visitadas,
            key =lambda ciudad: df.loc[actual,ciudad]
        )
        distancia_total += df.loc[actual,siguiente] # añade la distancia al contador

        ruta.append(siguiente)  # se añade a la ruta
        no_visitadas.remove(siguiente) # se elimina de no visitadas
        actual = siguiente  # se cambia la ciudad actual
    return ruta, distancia_total


ruta, distancia = calcular_ruta(df, "Vigo")
print('la ruta es: ',ruta)
print('La distancia total es: ',distancia, 'km')
```

(EXTRA) Des de quina ciutat la ruta seria més curta amb l'algoritme plantejat

```python
menor_distancia = float('inf') # ponemos infinito para que el primer valor sea inferior
mejor_ruta =  None   # para indicar que está vacío
mejor_ciudad = None

# se comprueba cada ciudad como origen y se calcula la ruta

for ciudad in df.columns:
    ruta, distancia = calcular_ruta(df, ciudad)

    if distancia < menor_distancia:
        menor_distancia = distancia
        mejor_ruta = ruta
        mejor_ciudad = ciudad


print("Mejor ciudad de salida:", mejor_ciudad)
print("Distancia total:", menor_distancia, 'km')
print("Ruta:", mejor_ruta)

```
