### 1. Consulta l’API pública JSONPlaceholder utilitzant el mètode GET per obtenir:

- Llista de publicacions (/posts)
- Llista d’usuaris (/users)
- Llista de tasques (/todos)

```python
import requests
import pandas as pd
import json
```

```python
base_url = "https://jsonplaceholder.typicode.com"

# Posts
posts = requests.get(f"{base_url}/posts")

# Users
users = requests.get(f"{base_url}/users")

# Todos
todos = requests.get(f"{base_url}/todos")


posts_json = posts.json()
users_json = users.json()
todos_json = todos.json()
```

### 2. Mostra per pantalla:

- La quantitat total de cada recurs.
- El codi d'estat de cada petició.

```python
print("POSTS")
print("Total:", len(posts_json))
print("Status:", posts.status_code)

print("\nUSERS")
print("Total:", len(users_json))
print("Status:", users.status_code)

print("\nTODOS")
print("Total:", len(todos_json))
print("Status:", todos.status_code)
```

### 3. Fes una petició a una publicació inexistent per obtenir un error 404 i mostra el codi d'estat rebut.

```python
error_response = requests.get(f"{base_url}/posts/101")
try:
    error_response.raise_for_status()
except requests.exceptions.HTTPError as e:
        print(f"Error HTTP: {e}")

```

### 4. Fes una petició POST per crear una nova publicació fictícia. Inclou el títol, el cos del missatge i un userId.

- Mostra la resposta JSON.
- Mostra el codi d'estat.

```python
new_post = {
    "title": "POST ficticio",
    "body": "Publicación de prueba",
    "userId": 1
}

response_post = requests.post(
    f"{base_url}/posts",
    json=new_post
)

print("Status:", response_post.status_code)
print(response_post.json())
```

### 5. Fes una petició PATCH per modificar parcialment una publicació existent.

- Mostra la resposta JSON.
- Mostra el codi d'estat.

```python
update_data = {
    "title": "Título modificado"
}

response_patch = requests.patch(
    f"{base_url}/posts/1",
    json=update_data
)

print("Status:", response_patch.status_code)
print(response_patch.json())
```

### 6. Fes una petició DELETE sobre una publicació.

- Mostra la resposta JSON.
- Mostra el codi d'estat.

```python
response_delete = requests.delete(f"{base_url}/posts/1")

print("Status:", response_delete.status_code)
print(response_delete.json())
```

# Nivell 3

### 1. Utilitza l’API d’Open Data BCN per:

Cercar un dataset del teu interès mitjançant package_search i package_show. .

```python
base = "https://opendata-ajuntament.barcelona.cat/data/api/3/action"

search = requests.get(f"{base}/package_search", params={
    "q": "transport"
})

results = search.json()
print(search.status_code)
print(len(results["result"]["results"]))
```

### 2. Dels resultats obtinguts, selecciona’n un que tingui recursos disponibles en CSV o JSON.

```python
dataset = results["result"]["results"][0]

dataset_info = requests.get(
    f"{base}/package_show",
    params={"id": dataset["id"]}
)

data_json = dataset_info.json()

resources = data_json["result"]["resources"]

for r in resources:
    print(r["name"], r["format"], r["id"])
```

### 3. Amb el resource_id del recurs seleccionat, realitza una consulta mitjançant datastore_search per:

Recuperar almenys 100 registres.

```python
resource_id = resources[0]["id"]

data = requests.get(
    f"https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search",
    params={
        "resource_id": resource_id,
        "limit": 100
    }
)

records = data.json()["result"]["records"]
    
    
```

### 4. Converteix els resultats a un DataFrame.

```python
df_bcn = pd.DataFrame(records)
df_bcn.head()
```

### 5. Desa el DataFrame en un fitxer .csv.

```python
df_bcn.to_csv("open_data_bcn.csv", index=False)
print("Guardado correctamente")
```
